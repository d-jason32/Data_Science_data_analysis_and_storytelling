## Data Cleaning Essentials

```python

# Missing values summary

df.isnull().sum()
```

```python
# Drop duplicates if any

df.drop_duplicates(inplace=True)
```

```python
# Fill missing numeric values with median

df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fill missing categorical values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
# Normalize text data
df['Sex']= df['Sex'].str.lower()
```

```python
df.info()
```

```python
#missing-value percentage check  more intuitive for EDA reports
(df.isnull().sum()/len(df)*100).sort_values(ascending=False)
```

```python
#Handling for incorrect data types
#Example: Convert 'Fare' to float if loaded as string
df['Fare']=df['Fare'].astype(float)
df
```

## Summary Statistics Review

```python

# Numerical summaries
print("Mean Age:", df['Age'].mean())
print("Std of Fare:",df['Fare'].std() )
print("Min & Max Fare:",df['Fare'].min(), df['Fare'].max() )

# Categorical summaries
print("\nPclass counts:")
print(df['Pclass'].value_counts())

#Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
print("\nEmbarked distribution:")
print(df['Embarked'].value_counts())
```

```python
#Another way of showing std and var
std = df[['Fare']]\
        .apply(['var', 'std'])
std
```

```python
#unique value counts and mode detection
print("\nUnique value counts per column:")
print(df.nunique())
```

## Exploring Relationships

```python
# Group-based summaries


print("Mean Fare by Passenger Class:")
print(df.groupby('Pclass')['Fare'].mean())

print("\nSurvival rate by Sex:")
print(df.groupby('Sex')['Survived'].mean())

print("\nSurvival rate by Class and Sex:")
print(df.groupby(['Pclass','Sex'])['Survived'].mean())

# Correlation matrix

df.corr(numeric_only=True)
```

```python
# Bivariate Analysis  Numeric vs Numeric
print("Correlation between Age and Fare:\n",df[['Age', 'Fare']].corr() )

# Bivariate Analysis  Numeric vs Categorical
print("\nAverage Fare by Survival Status:\n", df[['Fare', 'Survived']].corr())

# # Bivariate Analysis  Categorical vs Categorical
print("\nSurvival counts by Embarked port:\n",pd.crosstab(df['Survived'], df['Embarked']))
```

## Outlier Detection (Without Plots)

```python

# Using IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

outliers =df[(df['Fare']<(Q1-1.5*IQR))| (df['Fare']>(Q3+1.5*IQR))]
print("Number of outliers in Fare:", outliers.shape[0])

# Using Z-Score method
from scipy import stats
z = stats.zscore(df['Fare'])
print("Potential outliers (Z > 3):", df[abs(z) > 3].shape[0])
```
